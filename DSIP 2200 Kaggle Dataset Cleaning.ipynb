# 1,000 Meter Race Dataframe - 3 Sections

```python
#creating a dataframe for the one kilometer races
length_1000 = rrmerge.loc[rrmerge['race_distance'] == 1000]

#checking to see the contents of length_1000 dataframe
length_1000.shape
```

```python
#checking to see where NaN values are present within the dataframe
print(length_1000.isnull().sum())
```

##### As we can see above, there are 3,046 entires altogether and there exist 3,046 NaN values in running position 4, 5 and 6 each. Therefore, since we now know that all the values in those columns are NaN; I will just drop those columns as there is no confusion about whether we need them or not. There do exist 62 NaN values still present in running position 1, 2 and 3 columns however I will deal with them in the next section (Disallowed Section; see below)

```python
#dropping the running position columns that are not being used as we only need 3 sections for the length_1000 dataframe and 
length_1000 = length_1000.drop(['running_position_4', 'running_position_5', 'running_position_6'], axis=1)

length_1000
```

### Disallowed 1000 (da_1000)
##### I started by creating a dataframe called da_1000 to handle any entires that have NaN values from the running position 1 column for the 1000 meter races; This is important as it allows us to see what entires have no race data since having a first section running position has to exist in any race. However, it is worth noting that in these instances it do not mean that the race didn't happen, just that for one reason or another that particular horse was unfit to race or was disallowed. As an example, Shantaraam was disallowed to race on the veterinary officer's orders after it became irritated and reared on two separate occasions, more info at this link: https://racing.hkjc.com/racing/information/English/Racing/LocalResults.aspx?RaceDate=2014/09/14&Racecourse=ST&RaceNo=1

```python
#creating a dataframe of entries for the disallowed entires (These will be dropped)
da_1000 = length_1000[length_1000['running_position_1'].isnull()]

#creating a dataframe to check if there are values after running position 1
check_da_1000 = da_1000[da_1000['running_position_2'].notnull() | da_1000['running_position_3'].notnull()]

#checking to see the contents of this dataframe
check_da_1000.shape
```

##### There are 62 NaN values present in running position 1, 2 and 3 and we now know that da_1000 is a dataframe that contains all the remaining NaN values; I will remove them from the length_1000 dataframe as there is no confusion about whether they will be useful to us or not.

```python
#displaying the disallowed 1000 dataframe
da_1000
```

```python
#dropping NaN values from length_1000 dataframe as we can see above that all the disallowed frame NaN values
length_1000 = length_1000.dropna(subset = ['running_position_1'])

length_1000.shape
```

```python
#how many NaN values are present after cleaning?
print(length_1000.isnull().sum().sum())
```

### Did Not Finish 1000 (dnf_1000)
##### As an example; a horse I researched on the DNF list called "Snow Slider" is represented with a NaN value. A video of the race where he DNFs in the race can be located here; https://racing.hkjc.com/racing/video/play.asp?type=replay-full&date=20150207&no=01&lang=eng
##### Using the statements below I am checking to see if there are any records of horses that did not finish in the 1000 meter races. We already know from the disallowed section that there are no longer any NaN values in the dataset ands therefore no DNFs but I wanted to double check and as we can see below, there does not exist a single instance where a horse began the race and dropped out.

```python
#creating a dataframe to handle the did not finish entires
dnf_1000 = length_1000[length_1000['running_position_1'].notnull()]

#creating a dataframe to check if a horse started the race but didn't finish
check_dnf_1000 = dnf_1000[dnf_1000['running_position_2'].isnull() | dnf_1000['running_position_3'].isnull()]

#checking to see if there exists any horses that began the race but didn't finish
check_dnf_1000.shape
```

```python
#displaying the length_1000 dataframe in its final and cleaned state
length_1000.sample(3)
```
